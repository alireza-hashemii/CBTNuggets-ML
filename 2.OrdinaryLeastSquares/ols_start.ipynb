```python
import numpy as np
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
# Loading Dataset
Housing = datasets.fetch_california_housing()
print(Housing.feature_names)
```

```python
# Convert it to pandas dataframe
Housing_table = pd.DataFrame(
                    Housing.data, 
                    columns=Housing.feature_names,
                )
```

```python
# Feature Engineering 

Housing_table['MedInc'] = Housing_table['MedInc'] * 10000
Housing_table['HouseValue'] = Housing.target * 100000

Housing_table['MedInc'] = Housing_table['MedInc'].apply(float)
Housing_table['HouseValue'] = Housing_table['HouseValue'].apply(float)
```

```python
input_data = Housing_table['MedInc']
output_data = Housing_table['HouseValue']
```

```python
slope, intercept = np.polyfit(input_data.squeeze(),output_data,deg=1)
```

```python
House_Prediction_Formula = slope * input_data.squeeze() + intercept
```

```python
plt.scatter(input_data, output_data, color='blue', label='prices', alpha=.2)
plt.plot(input_data, House_Prediction_Formula, color='red', label='actual prices',alpha=.7)
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.legend()
plt.show()
```
