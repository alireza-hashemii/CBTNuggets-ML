```python
import seaborn as sns
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
dataset = sns.load_dataset('mpg')
dataset
```

```python
df = dataset.dropna()
x = df['horsepower'].values
y = df['mpg'].values

x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_test_reshaped = x_test.reshape(-1,1)
x_train_reshaped = x_train.reshape(-1,1)

model = LinearRegression()
model.fit(x_train_reshaped, y_train)
```

```python
slope = model.coef_[0]
y_intercept = model.intercept_
print(f"Slope is {slope}\nintercept is {y_intercept}")
```

```python
plt.scatter(x_train, y_train, label='Train Data')
plt.scatter(x_test, y_test, label='Test Data', color='r')
plt.plot(x_train, model.predict(x_train_reshaped),color='green', label=f'y_hat:{slope: .2f}x + {y_intercept: .2f}')
plt.xlabel('HorsePower')
plt.ylabel('Miles Per Gallon(MPG)')
plt.legend()
plt.show()
```

```python
y_pred = model.predict(x_test_reshaped)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f"""
Mean Absolute Error: {mae}
Mean Squared Error: {mse}
Root Mean Squared Error: {rmse}
""")
```
