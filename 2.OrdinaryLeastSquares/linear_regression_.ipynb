```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [2.75, 4, 3.85, 3, 4.8, 1.13, 
    2.3, 3.7, 5, 4.05,1.5,1.8,1,3.3,
]

y = [33, 66, 60, 44, 88, 11, 30,
    54, 99, 66.5, 25, 27, 6, 62
]

x_mean = sum(x) / len(x)
y_mean = sum(y) / len(y)
```

```python
numerator = 0
denominator = 0
for i in range(len(x)):
    numerator += (x[i] - x_mean) * (y[i] - y_mean)
    denominator += (x[i] - x_mean) ** 2

slope = numerator / denominator
y_intercept = y_mean - (slope * x_mean)

numpyar= np.array(x)
regression_formula = y_intercept + (slope * numpyar) 
```

```python
x_2 = np.arange(0, 10, 0.1)
w1=20
w0=-6
regression2 =w1*x_2+w0 
```

```python
plt.plot(x_2, regression2, 'g')
plt.scatter(x, y, marker='o')
plt.plot(x, regression_formula, 'r')
```
